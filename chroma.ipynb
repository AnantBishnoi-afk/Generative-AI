```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
docs = loader.load()
splitter  = RecursiveCharacterTextSplitter(chunk_size= 100,chunk_overlap=0)
splitdocs=splitter.split_documents(docs)
embeddings = OllamaEmbeddings(model="all-minilm")
```

```python
vectordb= Chroma.from_documents(splitdocs,embeddings)
```

```python
query = " father of nuclear material"
docs = vectordb.similarity_search(query)
print(docs[0].page_content)
```

```python
vectordb = Chroma.from_documents(documents=splitdocs,embedding=embeddings,persist_directory="./chroma-db")
```

```python
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
docs1 = db2.similarity_search(query)
print(docs1)
```

```python
retriever = vectordb.as_retriever()
result = retriever.invoke(query)
print(result[0].page_content)
```
