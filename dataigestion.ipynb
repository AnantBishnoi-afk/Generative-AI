```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://speakola.com/political/dr-a-p-j-abdul-kalam-vision-for-india-2011")
docs = loader.load()
print(docs)
```

```python
from bs4 import SoupStrainer
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(web_paths=["https://englishspeecheschannel.com/english-speeches/elon-musk-2021-speech/"],
                       bs_kwargs={ 'parse_only':SoupStrainer(class_=['post-title','post-content','post-header'])
                                 }
)
loader.load()
```

```python
from langchain_community.document_loaders import WikipediaLoader
loader = WikipediaLoader(query='Generative AI',load_max_docs=2)
docs = loader.load()
print(docs[0].page_content)
print(docs[0].metadata)
```
